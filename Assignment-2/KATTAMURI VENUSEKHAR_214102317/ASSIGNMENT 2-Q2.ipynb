```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
matrix1 = np.array([[0.5, 0],[0, 0.5]])
eig_val,eig_vec=np.linalg.eig(matrix1)
print('Eigen_values==>', eig_val)
print('Eigen_vector==>\n',eig_vec)
def mat1(x1, x2):
    return 0.5*(x1 ** 2 + x2 ** 2)
fig = plt.figure()
ax = plt.axes(projection='3d')
x1=np.linspace(-2,2,30)
x2=np.linspace(-2,2,30)
X1, Y1 = np.meshgrid(x1, x2)
Z= mat1(X1, Y1)
ax= plt.axes(projection ='3d')
surf=ax.contour3D(X1, Y1, Z, 50, cmap=cm.RdPu)
fig.colorbar(surf, shrink=0.5, aspect=10,cmap=cm.RdPu)
ax.set_title('')
plt.figure()
```

```python
matrix2 = np.array([[1, 0],[0,1]])
eig_val,eig_vec=np.linalg.eig(matrix2)
print('Eigen_values==>', eig_val)
print('Eigen_vector==>\n',eig_vec)
def mat2(x,y):
    return np.sin(np.sqrt(x **2 + y ** 2))
x=np.linspace(-8, 8, 30)
y=np.linspace(-8, 8, 30)
X,Y = np.meshgrid(x, y)
Z=mat2(X, Y)
fig = plt.figure()
ax = plt.axes()
ax.contour(X, Y, Z,cmap=cm.RdPu)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('contour plot')
```
