### BETA DISTRIBUTION

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

import warnings
warnings.filterwarnings('ignore')

def beta(x,alpha,bet):
    pdf =[]
    ga=math.gamma(alpha)
    gb=math.gamma(bet)
    gc=math.gamma(alpha+bet)
    den=(ga*gb)/gc
    for t in x:
        r=(t**(alpha-1))*((1-t)**(bet-1))/den
        pdf.append(r)
        
            
            
    cdf = np.cumsum(pdf)
    
    return pdf,cdf


t1=np.linspace(0,1,100)
lst = [beta(t1,1,3),beta(t1,5,1),beta(t1,2,5)]
p=['pdf','cdf']
c0=['black','blue','red']
fig,ax=plt.subplots(2,3,figsize=(8,6))
c=1
for i in range(len(lst)):
    for j in range(0,2): 
            plt.subplot(2,3,c)
            plt.plot(t1,lst[i][j])
           
            plt.title('plot of '+str(p[j]))
            c=c+1
```

PART A:
Consider a transmitter that can transmit messages only in the form
of 2 bits 0 and 1. The receiver receives a single message sent by the
transmitter say X. Since, we do not know before hand which bit
was sent X follows a certain distribution. Which distribution is this?
Find the PMF and plot it. Also calculate the mean and variance of
X. Assume probability for receiving 0(failure) is p = 0:1; 0:5; 0:8.

```python
import numpy as np 
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings('ignore')

def impulse(a, n,p): 
    delta =[] 
    for sample in n: 
        if sample == 0: 
            delta.append(p) 
        elif sample==1: 
            delta.append(1-p) 
            
        else:
            delta.append(0)
              
    return delta 

n=np.arange(-3,7,1)
prob=[0.1,0.5,0.8]
fig,ax=plt.subplots(3,1,figsize=(20,10))
for j,e in enumerate(prob,start=1):
    for i in range(2):
        res = impulse(i,n,e)
        plt.subplot(3,1,j)
        plt.stem(n,res,use_line_collection=True)
```

In the problem discussed above ,we have plotted the PMF for given three different values of P(probability of error) which is equivalent to receiving a zero at the output .So the PMF would be 2 impulses one at zero with amplitude equal to probability of error and other at one with amplitude = 1-probability of error.
This follows a bernoulli distribution .
Mean = p = probability of success
Variance=pq

for 1st case , mean=0.9,variance=0.09
for 2nd case , mean=0.5,variance=0.25
for 1st case , mean=0.2,variance=0.16

PART B

Consider the same transmitter but this time the receiver receives 20
bits. Each bit is independent of the previous one. The probability of
receiving a 0 is p = 0:4. Let X be the number of 1s (successes) that
is received at the receiver's end. What is the distribution followed
by X. How is it related to the rst distribution? Calculate the mean
and variance and plot the PMF.

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

import warnings
warnings.filterwarnings('ignore')

def nCr(n,r):
    
    return (fact(n)/(fact(r)*fact(n-r)))

def fact(n):
    
    res = 1
    for i in range(2,n+1):
        res = res * i
        
    return res

def binomial(t,n,p):
    pdf =[]
    
    for j in t:
        res = nCr(n,j)
        mul= ((1-p)**j)*(p**(n-j))
        bi=res*mul
        pdf.append(bi)
            
    return pdf


t=np.arange(5,20,1)
res = binomial(t,20,0.4)
plt.stem(t,res)
plt.title('PMF')
```

The PMF of X(success)= Receiving 1, follows binomial distribution that is nCr(1-p)^rp^(n-r). For varying values of X we have plotted a PMF of binomial distribution. Mean = np = 200 into 6=12 Variance= npq = 200 into 60 into 4=4.8

PART C

Consider an experiment in which bits are transmitted sequentially.
Each bit is independent of the other. When the rst bit arrives the
receiver checks for a 1. If it is 1 the receiver takes a decision. If
it is a 0 the receiver waits for the next bit. This goes on till the
receiver nds out that the bit was 1 and the experiment is dropped
once the bit is 0. Let X denote that the bit received was 1 on the kth
time (example, X = 2 suggests that the bit sequence was 0, 1 and
once 1 was received the experiment was stopped). Find the PMF
of X. What is this distribution known as? Calculate the mean and
variance and plot the PMF. How is this distribution related to the
rst distribution.

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

import warnings
warnings.filterwarnings('ignore')

def geometric(n,X,p):
    pdf =[]
    for t in n:
        if (t==X):
            pdf.append(1-p)
            
        else:
            pdf.append(p)
        
        
            
    return pdf


n=np.arange(1,11,1)
res = geometric(n,8,0.3)
plt.stem(n,res)
```
